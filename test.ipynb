```python
def moving_average(a, window_size):
    cumulative_sum = np.cumsum(np.insert(a, 0, 0)) 
    middle = (cumulative_sum[window_size:] - cumulative_sum[:-window_size]) / window_size
    r = np.arange(1, window_size-1, 2)
    begin = np.cumsum(a[:window_size-1])[::2] / r
    end = (np.cumsum(a[:-window_size:-1])[::2] / r)[::-1]
    return np.concatenate((begin, middle, end))
```

```python
lr = 1e-4
num_episodes = 1000
hidden_dim = 128*4
gamma = 0.98
epsilon = 0.01
target_update = 50
buffer_size = 500
minimal_size = 100
batch_size = 64
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
adjust_param = 1500

# env_name = 'Pendulum-v1'
# env = gym.make(env_name)
env_name = 'Fraud_Detection'
env = FraudDetectionEnv()
max_turns = len(env.train)//4
env = FraudDetectionEnv(max_turns)
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.n
```

### DQN + original

```python
replay_buffer = ReplayBuffer(buffer_size)
agent = CDQN(state_dim = state_dim,
             hidden_dim = hidden_dim,
             action_dim = action_dim,
             learning_rate = lr,
             gamma = gamma,
             epsilon = epsilon,
             target_update = target_update,
             device = device,
             dqn_type='DQN',
             q_estimate_type='original')
return_list, max_q_value_list, dqn_orig = train_DQN(agent, env, num_episodes, replay_buffer, 
                                                       minimal_size, batch_size)
```

```python
# save model
torch.save(dqn_orig, './trained_models/dqn_orig.pth')
```

```python
# load model
dqn_orig = torch.load('./trained_models/dqn_orig.pth')
```

```python
episodes_list = list(range(len(return_list)))
mv_return = moving_average(return_list, 5)
plt.plot(episodes_list, mv_return)
plt.xlabel('Episodes')
plt.ylabel('Returns')
plt.title('DQN_original on {}'.format(env_name))
plt.savefig('./figures/DQN_original on Returns')
plt.show()

frames_list = list(range(len(max_q_value_list)-adjust_param))
plt.plot(frames_list, max_q_value_list[adjust_param:])
# plt.axhline(0, c='orange', ls='--')
# plt.axhline(10, c='red', ls='--')
plt.xlabel('Frames')
plt.ylabel('Q value')
plt.title('DQN_original on {}'.format(env_name))
plt.savefig('./figures/DQN_original on Q value')
plt.show()
```

```python
true_positives_test = 0
false_positives_test = 0
true_negatives_test = 0
false_negatives_test = 0
accuracy = 0

for i in range(len(env.test)):
    action = dqn_orig.take_action(env.test.iloc[i,:-1].values)
    if int(action) == 1:
        if env.test.iloc[i,-1] == 1:
            true_positives_test += 1
        else:
            false_positives_test += 1
    elif int(action) == 0:
        if env.test.iloc[i,-1] == 0:
            true_negatives_test += 1
        else:
            false_negatives_test += 1
for i in range(len(env.test)):
    action = dqn_orig.take_action(env.test.iloc[i,:-1].values)
    if int(action) == env.test.iloc[i,-1]:
        accuracy += 1

accuracy_rate = accuracy/len(env.test)

fpr = false_positives_test / (false_positives_test + true_negatives_test)
tpr = true_positives_test / (true_positives_test + false_negatives_test)
recall = tpr
precision = true_positives_test / (true_positives_test + false_positives_test)
if not (precision+recall):
    f1 = 0
else:
    f1 = 2*precision*recall/(precision+recall)

print('FPR:',fpr)
print('TPR:',tpr)
print('Accuracy:',accuracy_rate)
print('Recall:',recall)
print('Precision:',precision)
print('F1-score:',f1)
```
